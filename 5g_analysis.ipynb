```python
def signedHexToInt(num):
  return -(num & 0x8000) | (num & 0x7fff)
```

```python
def readBin(path):
  file = open(path,"rb");
  fileSize = os.path.getsize(path);
  dataset =[];
  count = 0;
  previous = 0
  for line in file.read():
    if count == 0:
      previous = line;
      count += 1;
    else:
      count = 0;
      if (previous >= 16):
        dataset.append(signedHexToInt(int(hex(line) + hex(previous)[2:], 16)));
      else:
        dataset.append(signedHexToInt(int(hex(line)+'0'+hex(previous)[2:],16)));
  return dataset;
```

```python
def findStartPoint(wave):
  #para
  p1 = 0
  p2 = 0
  slope = 0
  position  = 0;
  #initializing
  for i in range(122880):
    if (i <= 61439):
      p1 += wave[i].real**2 + wave[i].imag**2
    else:
      p2 += wave[i].real**2 + wave[i].imag**2
  slope = p1 - p2

  #1st iteration
  for i in range(61440, 61440+228224+140416):
    p1 = p1 + wave[i].real**2 + wave[i].imag**2 - (wave[i - 30720].real**2 + wave[i - 30720].imag**2)
    p2 = p2 + wave[i + 30720].real**2+wave[i + 30720].imag**2 - (wave[i].real**2 + wave[i].imag**2)
    curSlope = p1 - p2
    if (curSlope > slope):
      slope = curSlope
      position = i

  #para
  p1 = 0
  p2 = 0
  position2 = 0
  slope2 = 0

  #initializing
  for i in range(position, position + 122880):
    if (i <= position + 61439):
      p1 += wave[i].real**2 + wave[i].imag**2
    else:
      p2 += wave[i].real**2 + wave[i].imag**2
  slope2 = p1 - p2

  #2nd iteration
  for i in range(position + 61440, position+61440+140416):
    p1 = p1 + wave[i].real**2 + wave[i].imag**2 - (wave[i - 61440].real**2 + wave[i - 61440].imag**2)
    p2 = p2 + wave[i + 61440].real**2+wave[i + 61440].imag**2 - (wave[i].real**2 + wave[i].imag**2)
    curSlope = p1 - p2
    if (curSlope < slope2):
      slope2 = curSlope
      position2 = i

  return [position, position2]

```

```python
def r(c):
  r = []
  for n in range(0, int(len(c)/2), 2):
    r.append(complex(1/math.sqrt(2)*(1- 2*c[n]),1/math.sqrt(2)*(1-2*c[n+1])));
  return r;
```
